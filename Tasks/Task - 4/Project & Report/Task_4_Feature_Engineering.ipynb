```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler, OneHotEncoder

# 1. Load the Dataset
# Loading the Adult Income Dataset from a standard repository
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/adult-all.csv"
columns = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status',
           'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss',
           'hours-per-week', 'native-country', 'income']

df = pd.read_csv(url, names=columns, na_values='?')

# Save raw file for GitHub
df.to_csv("adult_raw.csv", index=False)
print("--- RAW DATASET LOADED & SAVED ---")
display(df.head())

# Drop missing values for this task to keep it clean
df.dropna(inplace=True)

# 2. Identify Categorical and Numerical Features
numerical_cols = ['age', 'fnlwgt', 'education-num', 'capital-gain', 'capital-loss', 'hours-per-week']
categorical_cols = ['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'native-country']
target_col = 'income'

print(f"\nNumerical Columns: {len(numerical_cols)}")
print(f"Categorical Columns: {len(categorical_cols)}")

# 3. Scaling Numerical Features (StandardScaler)
# Scaling is crucial for algorithms that calculate distances (like KNN, SVM)
scaler = StandardScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

print("\n--- NUMERICAL FEATURES SCALED (StandardScaler) ---")
display(df[numerical_cols].head())

# 4. Feature Encoding

# A. Label Encoding (Target Variable & Ordinal Data)
# We apply this to the Target 'income' (>50K, <=50K)
le = LabelEncoder()
df[target_col] = le.fit_transform(df[target_col])
print(f"\nTarget Encoded: {le.classes_}")

# B. One-Hot Encoding (Nominal Data)
# We use get_dummies for categorical columns where order doesn't matter (e.g., Marital Status, Race)
# drop_first=True helps prevent multicollinearity (dummy variable trap)
df_processed = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

print("\n--- ONE-HOT ENCODING APPLIED ---")
print(f"New Dataframe Shape: {df_processed.shape}")
display(df_processed.head())

# 5. Save Processed Dataset
df_processed.to_csv("processed_adult_data.csv", index=False)
print("\nSUCCESS: 'processed_adult_data.csv' saved successfully!")
```
